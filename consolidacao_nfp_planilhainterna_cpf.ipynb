```python
import pandas as pd
import zipfile
import os

nome_zip = list(uploaded.keys())[0]

with zipfile.ZipFile(nome_zip, 'r') as z:
    z.extractall("pasta_dados")

arquivos = []
for root, dirs, files_list in os.walk("pasta_dados"):
    for file in files_list:
        if file.endswith(".xlsx") and not file.startswith("~$"):
            arquivos.append(os.path.join(root, file))

print(f"{len(arquivos)} arquivo(s) encontrado(s):")
for a in arquivos:
    print(" -", a)
```

```python
dfs = []
erros = []

for caminho in arquivos:
    try:
        df = pd.read_excel(caminho, sheet_name="CPF", engine="openpyxl")
        df["arquivo_origem"] = os.path.basename(caminho)
        dfs.append(df)
        print(f"✅ {os.path.basename(caminho)} — {len(df):,} linhas")
    except Exception as e:
        erros.append(caminho)
        print(f"❌ Erro em {os.path.basename(caminho)}: {e}")

consolidado = pd.concat(dfs, ignore_index=True)
print(f"\nTotal consolidado: {len(consolidado):,} linhas")

consolidado.to_excel("consolidado_cpf.xlsx", index=False)
print("Arquivo salvo!")
```
